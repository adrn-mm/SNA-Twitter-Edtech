```python
import pandas as pd
import collections
```

```python
df = pd.read_csv(r"c:\\Users\\LENOVO\\OneDrive\\Desktop\\Tugas Akhir\\Programming\\data\\raw/dataset.csv",
                 low_memory=False)
```

```python
df = df[["id","date","user_id","username","tweet","mentions",
         "urls","replies_count","retweets_count","likes_count","hashtags",
         "link","reply_to"]]
```

```python
df_ruangguru = df[df.username=="ruangguru"].copy()
df_zenius = df[df.username=="zeniuseducation"].copy()
df_ruangguru.sort_values(by=['replies_count','retweets_count','likes_count'],
                         ascending=[True, True, True],
                         na_position='first',ignore_index=True,inplace=True)
df_zenius.sort_values(by=['replies_count','retweets_count','likes_count'],
                         ascending=[True, True, True],
                         na_position='first',ignore_index=True,inplace=True)
df_zenius.reset_index(inplace=True, drop=True)
df_ruangguru.reset_index(inplace=True, drop=True)
```

```python
def my_count_hstg(df):
    lst_hshtgs = [eval(x) for x in df[df.hashtags != "[]"].hashtags]
    flt_hshtgs = [x for xs in lst_hshtgs for x in xs]
    counter=collections.Counter(flt_hshtgs)
    return counter.most_common()
[hstg_ruangguru, hstg_zenius] = [my_count_hstg(x) for x in [df_ruangguru, df_zenius]]
```
